```python
from eval.common import evaluate, evaluate_mAP
from generators.csv_ import CSVGenerator
from utils.anchors import anchors_for_shape
from layers import RegressBoxes
from model import efficientdet
import os

import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

os.environ['CUDA_VISIBLE_DEVICES'] = ''
```

```python
phi = 0
weighted_bifpn = True
common_args = {
    'batch_size': 1,
    'phi': phi,
}

test_generator = CSVGenerator(
    csv_data_file='/datasets/dataset/coco2017/annotations/instances_val2017_obb.csv',
    csv_class_file='/datasets/dataset/coco2017/annotations/classes_voc.csv',
    base_dir='/datasets/dataset/coco2017/val2017',
    **common_args
)
```

```python
model_path = 'checkpoints/gbb_l3/gbb_finetuned/csv.h5'
input_shape = (test_generator.image_size, test_generator.image_size)
anchors = test_generator.anchors
num_classes = test_generator.num_classes()
model, prediction_model = efficientdet(phi=phi, num_classes=num_classes, weighted_bifpn=weighted_bifpn)
prediction_model.load_weights(model_path, by_name=True)

i = 0
```

```python
''' LOAD IMAGE '''
x,y = test_generator[i]
i += 1

image = x[0][0].astype(np.float32)
image *= [0.229, 0.224, 0.225]
image += [0.485, 0.456, 0.406]
image = np.uint8(image*255)

''' GROUND TRUTH VALUES '''
anchor_state = y[1][:, :, -1]

# convert to boxes values: xmin, ymin, xmax, ymax
anchors = anchors_for_shape((test_generator.image_size, test_generator.image_size), anchor_params=None)
anchors_input = np.expand_dims(anchors, axis=0)
bt = RegressBoxes()([anchors_input, y[1][..., :4]])

# filter out "ignore" anchors
indices = tf.where(tf.keras.backend.equal(anchor_state, 1))
boxes_target  = tf.gather_nd(bt, indices).numpy()
angles_target = tf.gather_nd(y[1][..., -2], indices).numpy()*45.-45.

for box,angle in zip(boxes_target, angles_target):
    
    #xmin,ymin,xmax,ymax
    box = ((box[2]+box[0])/2, (box[3]+box[1])/2), ((box[2]-box[0]), (box[3]-box[1])), angle
    
    points = np.int0(cv2.boxPoints(box)) # 2x4 matrix, representing corner points
    _ = cv2.drawContours(image,[points], 0, (0,255,0), 2)
    
''' PREDICTED VALUES '''

boxes, scores, labels, angles = prediction_model.predict(x)
angles = (np.tanh(angles) - 1.)*45.

for score,box,angle in zip(scores[0], boxes[0], angles[0]):
    
    if score<0.3:
        continue
    
    #xmin,ymin,xmax,ymax
    box = ((box[2]+box[0])/2, (box[3]+box[1])/2), ((box[2]-box[0]), (box[3]-box[1])), angle
    
    points = np.int0(cv2.boxPoints(box)) # 2x4 matrix, representing corner points
    _ = cv2.drawContours(image,[points], 0, (255,0,0), 2)
    
plt.figure(figsize=(10,10))
plt.imshow(image)
```

```python
average_precisions = evaluate_mAP(test_generator, prediction_model, method='iou', visualize=False)

# compute per class average precision
total_instances = []
precisions = []
for label, (average_precision, num_annotations) in average_precisions.items():
    print('{:.0f} instances of class'.format(num_annotations), test_generator.label_to_name(label),
          'with average precision: {:.4f}'.format(average_precision))
    total_instances.append(num_annotations)
    precisions.append(average_precision)
mean_ap = sum(precisions) / sum(x > 0 for x in total_instances)
print('mAP0.5:0.95 = {:.4f}'.format(mean_ap))
```

```python
average_precisions = evaluate_mAP(test_generator, prediction_model, method='piou', visualize=False)

# compute per class average precision
total_instances = []
precisions = []
for label, (average_precision, num_annotations) in average_precisions.items():
    print('{:.0f} instances of class'.format(num_annotations), test_generator.label_to_name(label),
          'with average precision: {:.4f}'.format(average_precision))
    total_instances.append(num_annotations)
    precisions.append(average_precision)
mean_ap = sum(precisions) / sum(x > 0 for x in total_instances)
print('mAP0.5:0.95 = {:.4f}'.format(mean_ap))
```

## Visualize training

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
```

```python
def get_df(path_training, sufix='finetuned'):
    # open training info
    path_csv = glob(os.path.join(path_training+'*', '*.csv'), recursive=True)
    dfs = {'finetuned' if sufix in path else 'freezed': pd.read_csv(path) for path in path_csv}

    # freezed epochs
    freeze_epochs = max(dfs['freezed']['epoch'])
    dfs['finetuned']['epoch'] += freeze_epochs

    # join infos
    df = pd.concat([dfs['freezed'], dfs['finetuned']])
    df['epoch'] += 1
    return df
```

```python
def plot_metric(name, metric_name, df, diff=False):
    plt.figure(figsize=(10,5))
    plt.grid('on')
    if diff:
        from scipy.signal import savgol_filter
        adj = lambda x: savgol_filter(x[50:], 51, 3)
        epochs = df['epoch'][51:]
        plt.plot(epochs, adj(np.diff(df[f'val_{metric_name}loss'])),\
                 epochs, adj(np.diff(df[f'{metric_name}loss'])), linestyle='dashed')
        plt.legend(['val', 'train'])
        
    else:
        plt.plot(df['epoch'], df[f'val_{metric_name}loss'], df['epoch'], df[f'{metric_name}loss'])
        plt.axvline(x=50, color='r', linestyle='--')
        plt.legend(['val', 'train'])
    plt.xlabel('epoch')
    plt.ylabel(f'{metric_name}loss')
    plt.title(name)
    
    plt.savefig(f'saves/{metric_name}{name}.png')
```

```python
plot_metric('Smooth L1', 'classification_', df_l1)
plot_metric('GIoU', 'classification_', df_giou)
plot_metric('PIoU L1', 'classification_', df_pioul1)
plot_metric('PIoU L3', 'classification_', df_pioul3)
```
